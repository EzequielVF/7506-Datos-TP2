```python
import pandas as pd
import numpy as np

train_labels = pd.read_csv('Archivos de datos/train_labels.csv', encoding='latin-1')
train_values = pd.read_csv('Archivos de datos/train_values.csv', encoding='latin-1')
```

Me transformo las columnas tipo object en categoricos

```python
columnasCategoricas=np.array(['land_surface_condition', 'foundation_type', 'roof_type', 'ground_floor_type', 'position',\
                                    'plan_configuration', 'legal_ownership_status','other_floor_type'  
                                   ])
columnasCategoricas
```

```python
#Aplicamos get_dummies para librarnos de los categoricos con los que los modelos pudieran tener problemas

DumiesCat=  pd.get_dummies(train_values[columnasCategoricas])
train_values.drop(columns= columnasCategoricas,inplace=True)
train_values[DumiesCat.columns]=DumiesCat
train_values.info()
```

```python
train_values.to_csv('Archivos de datos/datos_procesados.csv',index=False)
```

```python
values=pd.read_csv('Archivos de datos/test_values_original.csv')
#Aplicamos get_dummies para librarnos de los categoricos con los que los modelos pudieran tener problemas


DumiesCat=  pd.get_dummies(values[columnasCategoricas])
values.drop(columns= columnasCategoricas,inplace=True)
values[DumiesCat.columns]=DumiesCat

values.to_csv('Archivos de datos/test_values_procesados.csv',index=False)
```

```python
values.info()
```

```python
#idea epicentro tomado TP1
train_values["epicentro"]=(train_values.geo_level_1_id==17).astype(int)+(train_values.geo_level_1_id==18).astype(int)
#idea materials tomado TP1
materials=np.array(['has_superstructure_adobe_mud','has_superstructure_mud_mortar_stone',
                          'has_superstructure_stone_flag','has_superstructure_cement_mortar_stone','has_superstructure_mud_mortar_brick',
                          'has_superstructure_cement_mortar_brick','has_superstructure_timber','has_superstructure_bamboo','has_superstructure_rc_non_engineered',
                          'has_superstructure_rc_engineered','has_superstructure_other'])
train_values['cant_matrials']=train_values[materials].sum(axis=1)
materiales_feos=['has_superstructure_stone_flag', 'has_superstructure_adobe_mud','has_superstructure_mud_mortar_brick']
train_values['materiales_feos']=train_values[materiales_feos].sum(axis=1)
materials_goods=['has_superstructure_rc_engineered','has_superstructure_cement_mortar_brick',       'has_superstructure_rc_non_engineered']
train_values['matial_goods']=train_values[materials_goods].sum(axis=1)
train_values.info()
```

```python
train_values.to_csv('Archivos de datos/Create_features_train.csv',index=False)
```
